```python
import random
#Fixed agent
def Agent1_Fixed():
    sum = 0
    for k in range(3):
        amount = (k+2)*10
        print ("Amount bid for round", k + 1, " is", amount)
        sum += amount
    return sum

#Random agent
def Agent2_Random():
    sum=0
    for j in range(3):
        amount=random.randint(0,51)
        print ("Amount bid for round",j+1," is",amount)
        sum+=amount
    return sum

# Function to identify the Hand
def identifyHand(Hand_):
    r=Hand_
    if (r[0] == r[1] and r[1] == r[2]):
        print ("player card is 3 of a kind")
    elif (r[0] == r[1] or r[0] == r[2] or r[1] == r[2]):
        print ("player card is  1 pair")
    else:
        print ("player card is  high card")
        
#function to decode the cards
def decodeCards(cards):
    d=[]
    for i in cards:
        if str.isdigit(i):
            d.append(int(i) - 1)
        elif (str.isalpha(i)):
            if i == 'T':
                d.append(9)
            elif i == 'J':
                d.append(10)
            elif i == 'Q':
                d.append(11)
            elif i == 'K':
                d.append(12)
            else:
                d.append(13)
    return d
# Function to analyse the hand
def strength_eval(Hand_):
    p=[]
    q=[]
    c=set()
    r=Hand_
    for x in r:
        if x in c:
            q.append(x)
        else:
            p.append(x)
            c.add(x)
    if (len(p)== 3):
        return (max(r)-2)
    elif(len(p)==2):
        if (q[0]==p[0]):
            return ((q[0]*10+p[1]))
        else:
            return ((q[0]*10+p[0]))
    else:
        return (q[0]*143)
#function to check the result of the game
def checkResult(r1,r2):
    if(r1>r2):
        print("The winner is PLAYER1")
        return 1
    elif(r2>r1):
        print ("The winner is PLAYER2")
        return 2
    else:
        return 3
#################################
#main function starts
#####################################3
point1=0
point2=0

for h in range(50):
    print ("Hand",h+1)
#card dealing phase
    r = ['2', '3', '4', '5', '6', '7', '8', '9', 'T', 'J', 'Q', 'K', 'A']
    s = ['s', 'h', 'c', 'd']
    shuffle = []
    Final = []
    while(len(Final)!=6):
        i = random.randint(0, 12)
        j = random.randint(0, 3)
        shuffle.append(r[i] + s[j])
        for x in shuffle:
            if x not in Final:
                Final.append(x)
                #print(Final)
    Hand_p1 = []
    Hand_p2 = []
    for x in range(6):
        if (x < 3):
         Hand_p1.append(Final[x]) 
         
        else:
             Hand_p2.append(Final[x])    
#To seperate ranks for both the players
    s = []
    r1 = []
    r2 = []
    for i in Final:
        for j in i:
            s.append(j)
    for k in range(len(s)):
        if k % 2 == 0:
            if (k <= 4):
                r1.append(s[k])
            else:
                r2.append(s[k])
                

#details of the player1
    print("Player1 details")
    print ("Cards of Player1=",Hand_p1)
    identifyHand(r1)
    d1 = decodeCards(r1)
    r1 = strength_eval(d1) 
#function call for fixed Agent-player1
    v1 = Agent1_Fixed()
    print ("Total amount bidded for this round is ", v1)
    print("---------------------------------")
#details of the player2
    print ("Player 2 details")
    print ("cards of Player1=",Hand_p2)
    identifyHand(r2)
    d2 = decodeCards(r2)
    r2 = strength_eval(d2)
#function call for Random Agent-player1
    v2=Agent2_Random()
    print ("Total amount bidded for this round is ",v2)
    print("-------------------------------")    
#check result
    res=checkResult(r1, r2)
    if(res==3):
       d1.sort(reverse=True)
       d2.sort(reverse=True)
       if(d1[1]==d2[1]):
           if(d1[2]>d2[2]):
               point1 += v1 + v2
               print ("Total points got for this round by Player1 is ", v1 + v2)
           else:
               point2 += v1 + v2
               print ("Total points got for this round by Player2 is ", v1 + v2)
       elif(d1[1]>d2[1]):
           point1 += v1 + v2
           print ("Total points got for this round by Player1 is ", v1 + v2)
       else:
           point2 += v1 + v2
           print ("Total points got for this round by Player2 is ", v1 + v2)
    if(res==1):
        point1+=v1+v2
        print ("Points got for the round by Player",res," is ", v1 + v2)
    elif(res==2):
        point2+=v1+v2
        print ("Points got for the round by Player",res," is ",v1 + v2)
    print ("================================")
#Final winner selection
print ("Total points obtained by Player1 is ",point1)
print ("Total points obtained by Player2 is ",point2)
if (point1>point2):
    print ("The final winner is Player1")
elif(point1<point2):
    print ("The final winner is Player2")
else:
    print ("It is a draw between two players")
    
```
