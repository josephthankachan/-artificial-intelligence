```python

import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.neural_network import MLPClassifier
from sklearn import tree

# import data
Data = np.loadtxt(open("Lab4Data.csv", "rb"), delimiter=";", skiprows=1)
print('Total Data length =', len(Data))

Train_set, Test_set = train_test_split(Data, test_size=0.2)
Input_train = Train_set[:, :8]     # input features
Target_train = Train_set[:, 9]  # output labels

Input_test = Test_set[:, :8]
Target_test = Test_set[:, 9]
print ('%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%')
print('Train Data length =', len(Train_set))
print('Test Data length =', len(Test_set))

##Case 1: Support_vector_Machine (SVM)
Support_vector_Machine = svm.SVR().fit(Input_train,Target_train)
print ('%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%')
print(Support_vector_Machine)

PredictedOutcome = Support_vector_Machine.predict(Input_test)
Number_of_Correct_Predictions = len([i for i, j in zip(PredictedOutcome, Target_test) if (i<j+1) and (i>j-1)])

print ('%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%')
print('Correct Predictions:', Number_of_Correct_Predictions, 'Out_of:', len(PredictedOutcome),
      'Test Data')
print('percentage of Accuracy Prediction =', (Number_of_Correct_Predictions/float(len(PredictedOutcome)))*100)

##Case 2: MLP
MLPR = MLPClassifier(hidden_layer_sizes=(15,), random_state=1, max_iter=1, warm_start=True).fit(Input_train,Target_train)
print ('%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%')
print(MLPR)

PredictedOutcome = MLPR.predict(Input_test)
Number_of_Correct_Predictions = len([i for i, j in zip(PredictedOutcome, Target_test) if (i<j+1) and (i>j-1)])

print ('%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%')
print('Correct Predictions:', Number_of_Correct_Predictions, 'Out_of:', len(PredictedOutcome),
      'Test Data')
print('percentage of Accuracy Prediction =', (Number_of_Correct_Predictions/float(len(PredictedOutcome)))*100)

##Case 3: Decision Tree
DCR = tree.DecisionTreeRegressor().fit(Input_train,Target_train)
print ('%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%')
print(DCR)

PredictedOutcome = DCR.predict(Input_test)
Number_of_Correct_Predictions = len([i for i, j in zip(PredictedOutcome, Target_test) if (i<j+1) and (i>j-1)])

print ('%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%')
print('Correct Predictions =', Number_of_Correct_Predictions, 'Out_of:', len(PredictedOutcome),
      'Test Data')
print('percentage of Accuracy Prediction =', (Number_of_Correct_Predictions/float(len(PredictedOutcome)))*100)
```
